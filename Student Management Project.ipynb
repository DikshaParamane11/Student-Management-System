```python
#Fixing subjects using Tuple
subjects = ("Math", "Science", "English")

#list for storing student data
students = []

#student add function(loop , dict , excep handling , functions)
def add_student():
    name = input("Enter student name: ")

    marks = {}
    for sub in subjects:
        while True:
            try:
                m = int(input(f"Enter marks for {sub}: "))
                marks[sub] = m
                break
            except:
                print("Invalid input! Enter numbers only.")

    student = {"name": name, "marks": marks}
    students.append(student)

    file = open("students.txt", "a")
    file.write(name + "," + ",".join(str(marks[s]) for s in subjects) + "\n")
    file.close()

    print("Student added successfully!\n")


#View student (File handling , loop)
def view_students():
    try:
        file = open("students.txt", "r")
        print("\n--- Student Records ---")
        for line in file:
            print(line.strip())
        file.close()
    except:
        print("No file found yet.\n")


#Result calculation(function , loop , strng)
def calculate_result():
    try:
        file = open("students.txt", "r")
        print("\n--- Results ---")
        for line in file:
            data = line.strip().split(",")
            name = data[0]
            marks = list(map(int, data[1:]))

            total = sum(marks)
            per = total / 3

            if per >= 75:
                grade = "Distinction"
            elif per >= 60:
                grade = "First Class"
            elif per >= 40:
                grade = "Pass"
            else:
                grade = "Fail"

            print(name, "->", per, "%", "-", grade)
        file.close()
    except:
        print("No student data available.\n")

#Topper finder(set + loop)
def show_topper():
    try:
        file = open("students.txt", "r")
        max_per = 0
        topper = ""

        for line in file:
            data = line.strip().split(",")
            name = data[0]
            marks = list(map(int, data[1:]))

            per = sum(marks) / 3
            if per > max_per:
                max_per = per
                topper = name

        file.close()
        print("\nTopper:", topper, "-", max_per, "%\n")
    except:
        print("No data found.\n")

#Menu system(loops)
while True:
    print("1. Add Student")
    print("2. View Students")
    print("3. Calculate Result")
    print("4. Show Topper")
    print("5. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        add_student()
    elif choice == "2":
        view_students()
    elif choice == "3":
        calculate_result()
    elif choice == "4":
        show_topper()
    elif choice == "5":
        print("Thank you")
        break
    else:
        print("Invalid choice!\n")
```
